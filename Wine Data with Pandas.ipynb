###  Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 
Assign it to a variable called wine

```python
import pandas as pd
import requests

# url="https://github.com/cs109/2014_data/blob/master/countries.csv"
# s=requests.get('wine.data').content
wine=pd.read_csv('wine.data')
wine  
```

# Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
# list_1 = [0,3,6,8,10,12,13]
wine = wine.drop(wine.columns[[0,3,6,8,10,12,13]], axis=1)
wine
```

## Step 5. Assign the columns as below:


1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine.columns = ['alcohol', 'malic_acid','alcalinity_of_ash','magnesium','flavanoids','proanthocyanins','hue']
wine
```

# Set the values of the first 3 rows from alcohol as NaN


```python
import numpy as np
wine['alcohol'][:3]=np.nan
wine
```

# Now set the value of the rows 3 and 4 of magnesium as NaNÂ¶

```python
wine['magnesium'][2:4]=np.nan
wine
```

# Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine['alcohol'] = wine['alcohol'].fillna(10)
wine['magnesium'] = wine['magnesium'].fillna(100)
wine
```
